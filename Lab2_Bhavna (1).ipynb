# Transfer Learning to classify a given sample image by using transfer learning using a pre-trained InceptionV3

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.inception_v3 import InceptionV3
from keras.applications.inception_v3 import preprocess_input, decode_predictions
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#Display Image
img = mpimg.imread('C://AI Sem2//Applied Machine Learning//sample_image.jpg')
imgplot = plt.imshow(img)
#load an image from file and reshaped
image = load_img('C://AI Sem2//Applied Machine Learning//sample_image.jpg', target_size=(299, 299))
#Convert the pixel image to a numpy array
image = img_to_array(image)
#reshape data for the model
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
#Prepare the image for the InceptionV3 model
image = preprocess_input(image)
#Load the model
model = InceptionV3(include_top=True)
#Predict the probability across all the output classes
yhat = model.predict(image)
#Convert the probabilities to class labels
label = decode_predictions(yhat)
#Retrieve the most likely result
label = label[0][0]
print(label[1])
print(label[2])
#Print the classification
print('%s (%.2f%%)' % (label[1], label[2]*100))
```
